```python
import jieba
from collections import Counter
```

```python
jieba.set_dictionary('./dict.txt.big')
```

```python
with open('./speech.txt', 'r', encoding = 'utf-8') as f:
    speech = f.read()
```

```python
"""
word_count = {'各位': 1,
              '台灣': 2,
              '民主': 999}
"""

word_count = Counter()
cur = jieba.cut(speech)
for word in cur:
    print(word)
    
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
```

```python
word_count.most_common(20)
```

```python
data_str = """
method:search
searchMethod:true
searchTarget:ATM
orgName:
orgId:
hid_1:1
tenderName:
tenderId:
tenderStatus:4,5,21,29
tenderWay:
awardAnnounceStartDate:105/12/22
awardAnnounceEndDate:105/12/22
proctrgCate:
tenderRange:
minBudget:
maxBudget:
_minBudget:
_maxBudget:
item:
hid_2:1
gottenVendorName:
gottenVendorId:
hid_3:1
submitVendorName:
submitVendorId:
location:
priorityCate:
isReConstruct:
btnQuery:查詢
"""
def gen_data(data_str):
    '''
    Convert key-vale strings to dict
    '''
    data_dict = {}
    for kv in data_str.strip().split('\n'):
        print(kv)
        k = kv.split(':')[0]
        v = kv.split(':')[1]
        data_dict[k] = v
    return data_dict
    
gen_data(data_str)
```
