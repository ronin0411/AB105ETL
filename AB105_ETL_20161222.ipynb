```python
import jieba
from collections import Counter
```

```python
jieba.set_dictionary('./dict.txt.big')
```

```python
with open('./speech.txt', 'r', encoding = 'utf-8') as f:
    speech = f.read()
```

```python
"""
word_count = {'各位': 1,
              '台灣': 2,
              '民主': 999}
"""

word_count = Counter()
cur = jieba.cut(speech)
for word in cur:
    print(word)
    
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
```

```python
word_count.most_common(20)
```
